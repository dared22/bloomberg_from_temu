### Task 1 : Find Top 5 markets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cloudscraper
```

```python
data = pd.read_excel("EQ_2025_06_30_Country.xlsx")
data
```

```python
data["Region"].unique()
```

```python
top_5_regions = data.groupby("Region")["Market Value(USD)"].sum().sort_values(ascending=False).head(5).to_frame("Market Value(USD)")

ax = sns.barplot(
    x=top_5_regions.index,
    y="Market Value(USD)",
    data=top_5_regions,
    palette="viridis",
    hue=top_5_regions.index
)
```

```python
top_5_countries = data.groupby("Country")["Market Value(USD)"].sum().sort_values(ascending=False).head(5).to_frame("Market Value(USD)")
ax = sns.barplot(
    x=top_5_countries.index,
    y="Market Value(USD)",
    data=top_5_countries,
    palette="viridis",
    hue=top_5_countries.index
)
```

```python
top_5_industries = data.groupby("Industry")["Market Value(USD)"].sum().sort_values(ascending=False).head(5).to_frame("Market Value(USD)")
plt.xticks(rotation=45, ha='right')
ax = sns.barplot(
    x=top_5_industries.index,
    y="Market Value(USD)",
    data=top_5_industries,
    palette="viridis",
    hue=top_5_industries.index
)
```
